# Understanding the Data

```python
features = ['done', 'distance', 'price', 'equipment', 'intercity', 'type', 'hurry', 'signup', 'expert']

fig, axes = plt.subplots(3, 3, figsize=(12, 11))

for i, feature in enumerate(features):
    ax = axes[i//3, i%3]  
    
    if feature == 'distance' or feature == 'price' or feature == 'signup':
        ax.hist(df[feature], bins=15, color='red', rwidth=0.4, alpha=0.6)
    else :
        ax.hist(df[feature], bins=5, color='blue', alpha=0.6)
    
    ax.set_xlabel(feature.capitalize())
    ax.set_ylabel('Count')
    ax.set_title(f'Distribution of {feature.capitalize()} Values')

plt.tight_layout()

plt.show()
```

#### Results:
1. Done Status: It is observed that approximately 75% of the requests were marked as "done", while about 25% were canceled.
2. Distance: The majority of customers are located near the service area, with distances close to zero. However, there is a noticeable subset of customers who are situated farther away.
3. Pricing: A significant amount of the requests had relatively low prices. But there were still few requests where the prices were considerably higher.
4. Equipment Requirement: The data reveals that almost all requests did not require additional equipment.
5. Intecity of Services: Nearly all requests were made within the same city as the service providers.
6. Service Type: Service type 'A' stands out as the most requested.
7. Emergency Requests: The data shows that most of the requests were not urgent or emergency-related.
8. Membership: The number of new members outweighs the number of older members.
9. Expert Group: There is a clear preference for expert type 'A'.

## Focal Loss:

```python
def focal_loss(alpha=0.1, gamma=2.5):
    def loss_fn(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1 - 1e-7)

        cross_entropy = -y_true * tf.math.log(y_pred) - (1 - y_true) * tf.math.log(1 - y_pred)
        focal_loss = alpha * tf.pow(1 - y_pred, gamma) * cross_entropy
        return tf.reduce_mean(focal_loss)
    
    return loss_fn
```

```python
def weighted_binary_crossentropy(class_0_weight, class_1_weight):
    def loss_fn(y_true, y_pred):
        bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
        
        weights = tf.where(tf.equal(y_true, 1), class_1_weight, class_0_weight)
        weighted_loss = weights * bce
        
        return tf.reduce_mean(weighted_loss)
    return loss_fn
```

## Model Compiler:

```python
def specificity(y_true, y_pred):
    tn = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 0), tf.float32))
    fp = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 1), tf.float32))
    specificity = tn / (tn + fp + tf.keras.backend.epsilon())
    return specificity

learning_rate = 0.0001
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

'''
model.compile(optimizer=optimizer, 
              loss='binary_crossentropy',
              #metrics=['accuracy', specificity])
              metrics=['accuracy', Precision()])
              #metrics=['accuracy', 'recall'])
'''

loss_fn = weighted_binary_crossentropy(3.0, 1.0)

model.compile(optimizer=optimizer, 
              #loss=loss_fn, 
              loss='binary_crossentropy',
              #loss=focal_loss(gamma=2.0, alpha=0.25),
              #metrics=['accuracy', Precision()])
              metrics=['accuracy', 'recall'])
```
