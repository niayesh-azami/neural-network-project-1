# Building Model

```python
x_train = train_df.drop('done', axis=1)
x_valid = valid_df.drop('done', axis=1)

x_train = x_train.drop('days_since_signup', axis=1)
x_valid = x_valid.drop('days_since_signup', axis=1)

x_train = x_train.drop('time', axis=1)
x_valid = x_valid.drop('time', axis=1)

y_train = train_df['done']
y_valid = valid_df['done']
```

```python
input_shape = [x_train.shape[1]]
input_shape
```

## Layrers:

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=input_shape),
    #tf.keras.layers.Dense(units=128, activation='relu'),
    #Dropout(0.5), 
    tf.keras.layers.Dense(units=64, activation='relu'),
    Dropout(0.5), 
    tf.keras.layers.Dense(units=32, activation='relu'),
    Dropout(0.5), 
    tf.keras.layers.Dense(units=16, activation='relu'),
    Dropout(0.5), 
    #tf.keras.layers.Dense(units=8, activation='relu'),
    #Dropout(0.5), 
    tf.keras.layers.Dense(units=4, activation='relu', kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.001)),
    Dropout(0.5), 
    #tf.keras.layers.Dense(units=2, activation='relu'),
    #Dropout(0.5), 
    tf.keras.layers.Dense(units=1, activation='sigmoid')  
])
model.summary()
```

## Focal Loss:

```python
def focal_loss(alpha=0.1, gamma=2.5):
    def loss_fn(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1 - 1e-7)

        cross_entropy = -y_true * tf.math.log(y_pred) - (1 - y_true) * tf.math.log(1 - y_pred)
        focal_loss = alpha * tf.pow(1 - y_pred, gamma) * cross_entropy
        return tf.reduce_mean(focal_loss)
    
    return loss_fn
```

```python
'''
def focal_loss(alpha_0, alpha_1, gamma=2.0):
    def loss_fn(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        alpha_t = y_true * alpha_1 + (1 - y_true) * alpha_0  # Use alpha values
        p_t = y_true * y_pred + (1 - y_true) * (1 - y_pred)
        focal_loss = -alpha_t * tf.pow(1 - p_t, gamma) * tf.math.log(p_t + tf.keras.backend.epsilon())
        return tf.reduce_mean(focal_loss)
    return loss_fn
    '''
```

## Model Compiler:

```python
def specificity(y_true, y_pred):
    tn = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 0), tf.float32))
    fp = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 1), tf.float32))
    specificity = tn / (tn + fp + tf.keras.backend.epsilon())
    return specificity

learning_rate = 0.0001
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

'''
model.compile(optimizer=optimizer, 
              loss='binary_crossentropy',
              #metrics=['accuracy', specificity])
              metrics=['accuracy', Precision()])
              #metrics=['accuracy', 'recall'])'''

model.compile(optimizer=optimizer, 
              #loss=focal_loss(alpha=0.75, gamma=2.0),
              loss='binary_crossentropy',
              #metrics=['accuracy', Precision()])
              metrics=['accuracy', 'recall'])
```

## Learning Rate:

```python
lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',  
    factor=0.5,          
    patience=2,          
    min_lr=1e-7         
    )
```

## Model Fit:

```python
'''class LearningRateLogger(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        lr = self.model.optimizer.learning_rate.numpy()
        print(f"Epoch {epoch + 1}: Learning Rate = {lr}")
'''
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5, 
    restore_best_weights=True
)
```

```python

history = model.fit(x_train, y_train,
    validation_data=(x_valid, y_valid),
    batch_size=32, 
    epochs=10,  
    #callbacks=[LearningRateLogger()]
    #callbacks=[lr_callback]
    #class_weight=class_weight
    callbacks=[lr_scheduler, early_stopping]
    )

```

## Plots:

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history.get('accuracy')  
val_acc = history.history.get('val_accuracy')  
train_pre = history.history.get('precision')  
val_pre = history.history.get('val_precision')  

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

if train_acc and val_acc:
    plt.subplot(1, 2, 2)
    plt.plot(train_acc, label='Training Accuracy')
    plt.plot(val_acc, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
plt.show()
```
