```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.preprocessing as StandardScaler
from sklearn.impute import KNNImputer
import seaborn as sns
from scipy.stats import zscore
#import tensorflow as tf
#import tensorflow.keras as keras
```

## Reading Data

```python
df = pd.read_csv('HW2_train.csv')
df.head(10)
```

## Understanding Data

### Done:

```python
unique_values = sorted(df['done'].unique())
bins = np.arange(min(unique_values) - 0.5, max(unique_values) + 1, 1)

plt.figure(figsize=(5, 5))
plt.hist(df['done'], bins=bins, color='#edbc7c', rwidth=0.4, alpha=0.7)

plt.xticks(unique_values)
plt.xlabel('done')
plt.ylabel('Count')
plt.title('Frequency of done Values')
plt.show()
```

### Distance:

```python
plt.figure(figsize=(8, 5))
plt.hist(df['distance'], bins=50, color='#92ccd7', alpha=0.7)

plt.xlabel('Distance')
plt.ylabel('Count')
plt.title('Distribution of Distance Values')
plt.show()
```

```python
days_df = df[df['distance'] < 0]['distance'].value_counts().reset_index(name='count')
days_df.columns = ['distance', 'count']
len(days_df)
```

### Price:

```python
plt.figure(figsize=(8, 5))
plt.hist(df['price'], bins=60, color='#92ccd7', alpha=0.7)

plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Price Values')
plt.show()
```

```python
days_df = df[df['price'] < 0]['price'].value_counts().reset_index(name='count')
days_df.columns = ['price', 'count']
len(days_df)
```

### Buying Equipment:

```python
unique_values = sorted(df['buying_equipment'].unique())
bins = np.arange(min(unique_values) - 0.5, max(unique_values) + 1, 1)

plt.figure(figsize=(5, 5))
plt.hist(df['buying_equipment'], bins=bins, color='#92ccd7', rwidth=0.4, alpha=0.7)

plt.xticks(unique_values)
plt.xlabel('Buying Equipment')
plt.ylabel('Count')
plt.title('Frequency of Buying Equipment Values')
plt.show()
```

```python
eq_df = df['buying_equipment'].value_counts().reset_index(name='count')
eq_df.columns = ['buying_equipment', 'count']
eq_df
```

### Intercity:

```python
unique_values = sorted(df['intercity'].unique())
bins = np.arange(min(unique_values) - 0.5, max(unique_values) + 1, 1)

plt.figure(figsize=(5, 5))
plt.hist(df['intercity'], bins=bins, color='#edbc7c', rwidth=0.4, alpha=0.7)

plt.xticks(unique_values)
plt.xlabel('Intercity')
plt.ylabel('Count')
plt.title('Frequency of Intercity Values')
plt.show()
```

```python
intercity_df = df['intercity'].value_counts().reset_index(name='count')
intercity_df.columns = ['intercity', 'count']
intercity_df
```

### Service Type:

```python
unique_values = sorted(df['service_type'].unique())

bins = np.arange(len(unique_values) + 1) - 0.5

service_type_numeric = df['service_type'].astype('category').cat.codes

plt.figure(figsize=(5, 5))
plt.hist(service_type_numeric, bins=bins, color='#8bc698', rwidth=0.4, alpha=0.7)

plt.xticks(range(len(unique_values)), unique_values)

plt.xlabel('Service Type')
plt.ylabel('Count')
plt.title('Frequency of Service Type Values')
plt.show()
```

```python
service_df = df['service_type'].value_counts().reset_index(name='count')
service_df.columns = ['service_type', 'count']
service_df
```

### In Hurry:

```python
unique_values = sorted(df['in_hurry_enabled'].unique())
bins = np.arange(min(unique_values) - 0.5, max(unique_values) + 1, 1)

plt.figure(figsize=(5, 5))
plt.hist(df['in_hurry_enabled'], bins=bins, color='#cf6464', rwidth=0.4, alpha=0.7)

plt.xticks(unique_values)
plt.xlabel('In Hurry Enabled')
plt.ylabel('Count')
plt.title('Frequency of In Hurry Enabled Values')
plt.show()
```

```python
hurry_df = df['in_hurry_enabled'].value_counts().reset_index(name='count')
hurry_df.columns = ['in_hurry_enabled', 'count']
hurry_df
```

### Days Since Signup:

```python
plt.figure(figsize=(8, 5))
plt.hist(df['days_since_signup'], bins=30, color='#92ccd7', alpha=0.7)

plt.xlabel('Days Since Signup')
plt.ylabel('Count')
plt.title('Distribution of Days Since Signup')
plt.show()
```

```python
days_df = df[df['days_since_signup'] < 0]['days_since_signup'].value_counts().reset_index(name='count')
days_df.columns = ['days_since_signup', 'count']
days_df
```

### Expert Group:

```python
df['expert_group'] = df['expert_group'].fillna('Unknown').astype(str)

unique_values = sorted(df['expert_group'].unique())

bins = np.arange(len(unique_values) + 1) - 0.5

expert_group_numeric = df['expert_group'].astype('category').cat.codes

plt.figure(figsize=(5, 5))
plt.hist(expert_group_numeric, bins=bins, color='#8bc698', rwidth=0.4, alpha=0.7)

plt.xticks(range(len(unique_values)), unique_values, rotation=45)

plt.xlabel('Expert Group')
plt.ylabel('Count')
plt.title('Frequency of Expert Group Values')
plt.show()
```

```python
exp_df = df['expert_group'].value_counts().reset_index(name='count')
exp_df.columns = ['expert_group', 'count']
exp_df
```

### Compare Done And Canceled Requests In Features:

```python
features = ['distance', 'price', 'days_since_signup']  # Replace with your actual feature names

# Set up the plotting grid
plt.figure(figsize=(12, len(features) * 4))

for i, feature in enumerate(features, 1):
    plt.subplot(len(features), 1, i)
    sns.kdeplot(data=df[df['done'] == 1], x=feature, fill=True, label='done = 1', color="blue", alpha=0.5)
    sns.kdeplot(data=df[df['done'] == 0], x=feature, fill=True, label='done = 0', color="red", alpha=0.5)
    plt.title(f'Density Plot of {feature}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# List of features with discrete values
discrete_features = ['buying_equipment', 'service_type', 'intercity']  # Replace with your actual feature names

plt.figure(figsize=(12, len(discrete_features) * 4))

for i, feature in enumerate(discrete_features, 1):
    plt.subplot(len(discrete_features), 1, i)
    sns.histplot(data=df[df['done'] == 1], x=feature, color="blue", alpha=0.5, label='done = 1', stat='count', bins=10)
    sns.histplot(data=df[df['done'] == 0], x=feature, color="red", alpha=0.5, label='done = 0', stat='count', bins=10)
    plt.title(f'Histogram of {feature}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python

# Assuming 'datetime' is already in datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Update the palette dictionary to use integers
plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='datetime', hue='done', palette={1: 'blue', 0: 'red'})
plt.title('Count of Occurrences Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title='Done', labels=['done = 0', 'done = 1'])
plt.tight_layout()
plt.show()
```
