## Filling The Null Values

```python
fill_df = df[["in_hurry_enabled", "service_type", "expert_group"]]
fill_df = pd.get_dummies(fill_df, columns=['service_type'])

non_null_df = fill_df[fill_df["expert_group"].notna()]
null_df = fill_df[fill_df["expert_group"].isna()]

x_non_null = non_null_df.drop(columns = ["expert_group"])
y_non_null = non_null_df["expert_group"].astype("category").cat.codes

categories = non_null_df["expert_group"].astype("category").cat.categories

x_null = null_df.drop(columns = ["expert_group"])

y_non_null = y_non_null.astype("category").cat.categories

imputer = KNNImputer(n_neighbors = 2)
imputer.fit(x_non_null)

x_null_imputed = imputer.transform(x_null)

null_df = null_df.copy()

null_df["expert_group"] = pd.Categorical.from_codes(
    x_null_imputed[:, -1].round().astype(int), categories
)

#null_df.loc[:, "expert_group"] = x_null_imputed[:, -1]

final_df = pd.concat([non_null_df, null_df])

final_df
```

### Merging Two Tables

```python
df.drop(columns = "service_type", inplace = True)
df.drop(columns = "expert_group", inplace = True)
df
```

```python
df = df.merge(final_df, left_index = True, right_index = True)
```

## Outlier Detection

#### The feature that might contain outliers are "distance", "price", and "days_since_signup".

### Distance Outliers

```python
sns.boxplot(x=df['distance'])
```

```python
sns.histplot(df['distance'])
```

```python
df['z_score'] = zscore(df['distance'])
```

```python
for threshold in [2, 2.5, 3, 3.5, 4, 4.5]:
    outliers = df['z_score'].abs() > threshold
    num_outliers = outliers.sum()
    print(f'Threshold: {threshold}, Outliers Detected: {num_outliers}')
```

#### Based on the result we will set the threshold to 3.

```python
threshold = 3
df['is_outlier'] = df['z_score'].abs() > threshold
```

```python
sorted_df = df.sort_values(by = 'distance')
sorted_df
```

```python
cap_value = df['distance'].quantile(0.95)
df['distance'] = np.where(df['is_outlier'], cap_value, df['distance'])
```

```python
check_df = df[df['is_outlier'] == True]
check_df
```

#### Invalid Distances (Negative Distances):

```python
positive_mean = df[df['distance'] > 0]['distance'].mean()
```

```python
df['distance'] = df['distance'].apply(lambda x: positive_mean if x < 0 else x)
```

```python
sns.histplot(df['distance'])
```
