# Outlier Detection On Train Set
## (Should Be Checked After Training)

## Distance:

```python
sns.boxplot(x=train_df['distance'], color = '#4340ff')
plt.show()
```

```python
column = 'distance'

Q1 = train_df[column].quantile(0.25)
Q3 = train_df[column].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.2 * IQR

outliers = train_df[(train_df[column] > upper_bound)]
inliers = train_df[(train_df[column] <= upper_bound)]

print(f"Number of outliers: {len(outliers)}")
print(f"Number of inliers: {len(inliers)}")

df_cleaned = inliers 
```

```python
sns.boxplot(x=df_cleaned['distance'], color = '#4340ff')
plt.show()
```

## Price:

```python
sns.boxplot(x=df_cleaned['price'], color = '#4340ff')
plt.show()
```

```python
column = 'price'

Q1 = df_cleaned[column].quantile(0.25)
Q3 = df_cleaned[column].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.4 * IQR

outliers = df_cleaned[(df_cleaned[column] > upper_bound) | (df_cleaned[column] < lower_bound)]
inliers = df_cleaned[(df_cleaned[column] <= upper_bound) & (df_cleaned[column] >= lower_bound)]

print(f"Number of outliers: {len(outliers)}")
print(f"Number of inliers: {len(inliers)}")

df_cleaned1 = inliers 
```

```python
sns.boxplot(x=df_cleaned1['price'], color = '#4340ff')
plt.show()
```
