```python
'''model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=input_shape),
    tf.keras.layers.Dense(units=16, activation='relu'),
    tf.keras.layers.Dense(units=16, activation='relu'),
    #tf.keras.layers.Dense(units=16, activation='relu'),
    #tf.keras.layers.Dense(units=64, activation='sigmoid'),
    tf.keras.layers.Dense(units=1)
])
model.summary()'''
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=input_shape),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')  # Use sigmoid for binary classification
])
model.summary()
```

```python
'''#model.compile(optimizer='adam', loss='mae')  
model.compile(optimizer='adam', 
              loss='binary_crossentropy',  # Or another loss function depending on your task
              metrics=['accuracy', Precision()])
'''

def specificity(y_true, y_pred):
    tn = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 0), tf.float32))
    fp = tf.reduce_sum(tf.cast((y_true == 0) & (tf.round(y_pred) == 1), tf.float32))
    specificity = tn / (tn + fp + tf.keras.backend.epsilon())
    return specificity

model.compile(optimizer='adam', 
              loss='binary_crossentropy',
              #metrics=['accuracy', Precision()])
              metrics=['accuracy', 'recall'])
```
